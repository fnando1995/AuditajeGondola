```python
import pandas as pd
import json
def printdata(hall):
    GT = pd.read_csv(f"datasets\evaluacion\\annotations\clusters\\2_full.csv",dtype=str,sep=",")
    GT = GT[GT["hall"]==hall]
    GT = list(set(GT["DESCRIPCION"]))
    print(len(GT))
    with open(f"datasets\evaluacion\positional_information\\4_pred_stage3_hall_{hall}.json") as f:
        D=json.load(f)
    DESC = list(set([d["best"][3][0] for c,d in D.items()]))
    print(len(DESC))
    print(" ========= in DESC but not in GT ========= ")
    for i in DESC:
        if i not in GT:
            print(i)
    print(" ========= in GT but not in DESC ========= ")
    for i in GT:
        if i not in DESC:
            print(i)
```
