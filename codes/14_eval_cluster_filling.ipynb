```python
import pandas as pd
import glob
import os


df = pd.read_csv("datasets\evaluacion\\annotations\clusters\eval.csv",sep=",",
dtype={
"hall":str,"cluster":str,"price":str,"code":str,"description":str,"base_price":str
})

maestra = pd.read_csv("datasets\evaluacion\maestra_filtered.csv",sep=",",
dtype={
"SUCURSAL":str,"code":str,"DESCRIPCION":str,"PVP":str
})

```

```python
dff=pd.merge(
    df,
    maestra,
    how="inner",
    on="code")[["hall","cluster","PVP","price","code","DESCRIPCION"]]
```

```python
dff.sort_values(by=['hall', 'cluster'],axis=0,inplace=True,ignore_index=True)
```

```python
dff.to_csv(
       "datasets\evaluacion\\annotations\clusters\eval.csv"
       ,sep=","
       ,index=False
   )
```
